# MAML-Select Resubmission Figures
Run the remote suite first. This notebook regenerates the EPS-only publication exports and lists the manuscript figures.

```python
from pathlib import Path
import subprocess
import sys
import pandas as pd
from IPython.display import display
from csfl_simulator.experiments.maml_select.run_experiments import REPO_ROOT

RESULTS = REPO_ROOT / 'runs' / 'maml_select'
ARTIFACTS = REPO_ROOT / 'artifacts' / 'maml_select'
ANALYSIS = ARTIFACTS / 'analysis'
PLOTS = ARTIFACTS / 'plots'
subprocess.run([sys.executable, '-m', 'csfl_simulator.experiments.maml_select.analyze_results', '--results-dir', str(RESULTS), '--output-dir', str(ANALYSIS)], check=True)
subprocess.run([sys.executable, '-m', 'csfl_simulator.experiments.maml_select.generate_plots', '--results-dir', str(RESULTS), '--output-dir', str(PLOTS)], check=True)
print(f'Figures written to {PLOTS}')
```

```python
summary = pd.read_csv(ANALYSIS / 'main_summary.csv')
tests = pd.read_csv(ANALYSIS / 'paired_significance_tests.csv')
display(summary.head(30))
display(tests.head(50))
```
