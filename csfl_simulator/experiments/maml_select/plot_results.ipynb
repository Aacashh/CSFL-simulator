# MAML-Select Resubmission Figures
Run the remote suite first. This notebook regenerates the EPS/PDF/PNG exports and previews the manuscript figures.

```python
from pathlib import Path
import subprocess
import sys
import pandas as pd
from IPython.display import Image, display
from csfl_simulator.core.utils import ART_ROOT

RESULTS = ART_ROOT / 'maml_select_letter'
ANALYSIS = RESULTS / 'analysis'
subprocess.run([sys.executable, '-m', 'csfl_simulator.experiments.maml_select.analyze_results', '--results-dir', str(RESULTS)], check=True)
print(f'Figures written to {ANALYSIS}')
```

```python
summary = pd.read_csv(ANALYSIS / 'main_summary.csv')
tests = pd.read_csv(ANALYSIS / 'paired_significance_tests.csv')
display(summary.head(30))
display(tests.head(50))
```
